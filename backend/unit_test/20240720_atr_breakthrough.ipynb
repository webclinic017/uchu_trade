```python
import yfinance as yf
import pandas as pd
import talib as ta

from _data_center.data_gather.ticker_price_collector import TickerPriceCollector
```

```python
# 下载BTC-USD数据
btc_data = TickerPriceCollector('BTC-USD').get_ticker_price()

# 计算3日ATR
btc_data['ATR_3'] = ta.ATR(btc_data['High'], btc_data['Low'], btc_data['Close'], timeperiod=3)

# 计算开多价格（收盘价 + 3日ATR的40%）
btc_data['Long_Entry_Price'] = btc_data['Close'] + 0.4 * btc_data['ATR_3']

# 提取最近几天的数据进行展示
btc_data
```
